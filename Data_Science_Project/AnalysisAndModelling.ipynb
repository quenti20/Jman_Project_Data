```python
from sklearn.preprocessing import LabelEncoder
from scipy.stats import chi2_contingency, f_oneway

# Create a copy of the DataFrame for preprocessing
df_encoded = df.copy()

# Initialize LabelEncoder
label_encoder = LabelEncoder()

# Apply label encoding to categorical variables
categorical_cols = ['TRAINING_NAME', 'TRAINER_NAME', 'USERTYPE', 'STATUS']
for col in categorical_cols:
    df_encoded[col] = label_encoder.fit_transform(df[col])

# Chi-square test for categorical variables
for col in categorical_cols:
    crosstab = pd.crosstab(df_encoded[col], df_encoded['DURATION'])
    chi2, p, dof, expected = chi2_contingency(crosstab)
    print(f"Chi-square test results for {col}:")
    print(f"  Chi-square statistic: {chi2}")
    print(f"  p-value: {p}")

# ANOVA test for numerical variables
numerical_cols = ['DURATION']
for col in numerical_cols:
    groups = [df_encoded[df_encoded[col] == val]['DURATION'] for val in df_encoded[col].unique()]
    f_statistic, p_value = f_oneway(*groups)
    print(f"ANOVA test results for {col}:")
    print(f"  F-statistic: {f_statistic}")
    print(f"  p-value: {p_value}")
```

```python
from scipy.stats import f_oneway

# Define the categorical variables
categorical_cols = ['TRAINING_NAME', 'TRAINER_NAME', 'USERTYPE', 'STATUS']

# Perform ANOVA for each categorical variable
for col in categorical_cols:
    # Group 'DURATION' values by the categories of the current categorical variable
    groups = [df[df[col] == category]['DURATION'] for category in df[col].unique()]
    
    # Perform ANOVA
    f_statistic, p_value = f_oneway(*groups)
    
    # Print ANOVA results
    print(f"ANOVA test results for {col}:")
    print(f"  F-statistic: {f_statistic}")
    print(f"  p-value: {p_value}")
```
