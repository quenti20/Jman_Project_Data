```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('data.csv')
df.head()
```

```python
df['DURATION'] = df['DURATION'].str.split().str[0]
df['DURATION']
```

```python
df['TRAINING_END_TIME'] = pd.to_datetime(df['TRAINING_END_TIME'])
df['TRAINING_START_TIME'] = pd.to_datetime(df['TRAINING_START_TIME'])

df['START_DATE'] = df['TRAINING_START_TIME'].dt.date
df['END_DATE'] = df['TRAINING_END_TIME'].dt.date
```

```python
col = ['TRAINING_START_TIME', 'TRAINING_END_TIME','WORKSESSION_ID']

# Check if each column in 'col' is present in the DataFrame's columns
if all(c in df.columns for c in col):
    # Drop the columns specified in 'col' from the DataFrame
    df = df.drop(columns=col, axis=1)

df.head()
```

```python
df['DURATION'].astype(float)
```

```python
df.head()
```

```python
from sklearn.preprocessing import OneHotEncoder

cols = ['USERTYPE_Employee','USERTYPE_Intern','STATUS_Completed','STATUS_Not Completed']

# Check if all elements in cols are in the DataFrame columns
if not all(col in df.columns for col in cols):
    one_hot_userType = pd.get_dummies(df, columns=['USERTYPE'])
    df = pd.concat([df, one_hot_userType], axis=1)
    df['USERTYPE_Employee'] = df['USERTYPE_Employee'].astype(int)
    df['USERTYPE_Intern'] = df['USERTYPE_Intern'].astype(int)
    
    one_hot_status = pd.get_dummies(df, columns=['STATUS'])
    df = pd.concat([df, one_hot_status], axis=1)
    df['STATUS_Completed'] = df['STATUS_Completed'].astype(int)
    df['STATUS_Not Completed'] = df['STATUS_Not Completed'].astype(int)
    df = df.iloc[:, :-2]
    
df.head()
```

```python
df = df.loc[:, ~df.columns.duplicated()]
```

```python
# df.drop(columns = ['STATUS_Completed','STATUS_Not Completed'],inplace = True)
# df.head()
```

```python
from sklearn.preprocessing import LabelEncoder

# Check if 'TRAINING_NAME_ENCODED' and 'TRAINER_NAME_ENCODED' columns already exist
if 'TRAINING_NAME_ENCODED' not in df.columns and 'TRAINER_NAME_ENCODED' not in df.columns:
    # Initialize label encoder
    encoder = LabelEncoder()

    # Label encode the TRAINING_NAME column
    df['TRAINING_NAME_ENCODED'] = encoder.fit_transform(df['TRAINING_NAME'])
    df['TRAINER_NAME_ENCODED'] = encoder.fit_transform(df['TRAINER_NAME'])

df.head()
df['TRAINER_NAME_ENCODED'].value_counts()
```

```python
#df.drop(columns = 'TRAINING_NAME_ENCODED',inplace=True)
df.head()
```

```python
dropped_cols = ['START_DATE','END_DATE','TRAINER_NAME','DURATION','USERTYPE','STATUS','TRAINING_NAME']
# Split data into features (X) and target variable (y)
X = df.drop(columns = dropped_cols, axis=1)  # Excluding 'START_DATE' and 'END_DATE'
y = df['DURATION']
X.head()
```

```python
import autosklearn
print('autosklearn: %s' % autosklearn.__version__)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error,r2_score
from autosklearn.regression import AutoSklearnRegressor
from autosklearn.metrics import mean_absolute_error,mean_squared_error
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
```

```python
print(X_train)
```

```python
automl = AutoSklearnRegressor(time_left_for_this_task=120, per_run_time_limit=30, n_jobs=-1)

# Fit the model
automl.fit(X_train, y_train)

# Get the models' performance statistics
y_hat = automl.predict(X_test)
mae = mean_absolute_error(y_test, y_hat)
r2 = r2_score(y_test, y_hat)
mse = mean_squared_error(y_test, y_hat)

print("Mean Absolute Error (MAE): %.3f" % mae)
print("R-Squared (R2): %.3f" % r2)
print("Mean Squared Error (MSE): %.3f" % mse)
```

```python
from sklearn.linear_model import Lasso
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error

# Initialize Lasso Regression model
lasso_model = Lasso(alpha=0.1)

# Fit the model to the training data
lasso_model.fit(X_train, y_train)

# Predict on the testing set
y_pred = lasso_model.predict(X_test)

# Calculate R-squared score
r2 = r2_score(y_test, y_pred)

# Calculate Mean Squared Error (MSE)
mse = mean_squared_error(y_test, y_pred)

# Calculate Mean Absolute Error (MAE)
mae = mean_absolute_error(y_test, y_pred)

print("R-squared Score:", r2)
print("Mean Squared Error:", mse)
print("Mean Absolute Error:", mae)
```
